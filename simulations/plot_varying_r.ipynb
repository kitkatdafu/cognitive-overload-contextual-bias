```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams["font.family"] = "Palatino"
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"

label_size = 22
tick_size = 20
title_size = 24
legend_size = 14
```

```python
df = pd.read_parquet("data/varying_r.parquet")
df.rename({'num_images_per_query': 'm'}, axis=1, inplace=True)
df.head(2)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
sns.lineplot(data=df, x="r", y="VI", hue="m", style="m", ax=ax, err_style="bars", markers=True)
ax.set_ylabel("VI", fontsize=label_size)
ax.set_xlabel("Percentage of edges observed", fontsize=label_size)
ax.tick_params(axis='both', which='major', labelsize=tick_size)
ax.legend(fontsize=legend_size)
fig.savefig("plots/simulation_vi_varying_r.pdf", bbox_inches="tight")
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
sns.lineplot(data=df, x="r", y="edge_error_rate", hue="m",
             style="m", ax=ax, err_style="bars", markers=True)
ax.set_ylabel("edge error rate", fontsize=label_size)
ax.set_xlabel("Percentage of edges observed", fontsize=label_size)
ax.tick_params(axis='both', which='major', labelsize=tick_size)
ax.legend(fontsize=legend_size)
fig.savefig("plots/simulation_edge_error_rate_varying_r.pdf", bbox_inches="tight")
```
