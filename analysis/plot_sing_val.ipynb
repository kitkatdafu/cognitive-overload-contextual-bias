```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
for dataset in ['dogs3', 'birds5']:
    for method in ['sd-rq-wp', 'radio']:
        for m in range(2, 9):
            if method == 'radio' and m > 4:
                continue
            if method == 'radio':
                filename = f'radio-{m}-image-{dataset}-2023.mat'
                labelname = 'radio'
            else:
                filename = f'sd-rq-wp-{m}-{dataset}-2023.mat'
                labelname = 'drag-and-drop'
            mat = loadmat('mats/' + filename)
            A = mat['adjacency_matrix']
            _, s, _ = np.linalg.svd(A)
            ax = plt.gca()
            ax.plot(np.arange(len(s)), s, label=f'{labelname}-{m}-{dataset}', marker='o')
            ax.set_xlabel('Singular Value Index')
            ax.set_ylabel('Singular Value')
            ax.legend()
            plt.savefig(f'plots/sing_val/{method}-{m}-{dataset}.png')
            plt.show()
```
