```python
import numpy as np
import seaborn as sns
from scipy.io import loadmat
import matplotlib.pyplot as plt
```

```python
num_obs, count = np.unique(
    loadmat("sd-rq-wp-3-dogs3-2023.mat")["adjacency_frequency_matrix"].reshape(-1),
    return_counts=True,
)
fig, ax = plt.subplots(figsize=(10, 5))
num_obs, count = num_obs[1:], count[1:]
sns.barplot(x=num_obs, y=count, ax=ax)
percentage = count / count.sum() * 100

patches = ax.patches
for i in range(len(patches)):
    x = patches[i].get_x() + patches[i].get_width() / 2
    y = patches[i].get_height() + 0.2
    ax.annotate("{:.1f}%".format(percentage[i]), (x, y), ha="center")
```
